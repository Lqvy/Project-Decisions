```python
# 1. Import Libraries & Settings

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn style for better visuals
sns.set(style="whitegrid", palette="Set2", context="notebook")
plt.rcParams['figure.figsize'] = (8,5)  # default figure size
```

```python
# 2. Load Diabetes Dataset

diabetes = pd.read_csv("diabetes_012_health_indicators_BRFSS2015.csv")

# Quick look
print("First 5 rows:\n", diabetes.head())
print("\nInfo:\n")
print(diabetes.info())
print("\nSummary statistics:\n")
print(diabetes.describe())
```

```python
# 3. Check for Missing Values

missing_vals = diabetes.isna().sum()
print("\nMissing values per column:\n", missing_vals)
```

```python
# 4. Histograms of Numeric Features

numeric_vars = diabetes.select_dtypes(include=np.number)

numeric_vars.hist(bins=30, figsize=(15,12), color='steelblue', edgecolor='black')
plt.suptitle("Numeric Feature Distributions", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# 5. Correlation Heatmap

plt.figure(figsize=(12,10))
corr_matrix = numeric_vars.corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```

```python
# 6. Boxplots: Diabetes_012 vs Numeric Features

for col in numeric_vars.columns:
    if col != 'Diabetes_012':
        plt.figure(figsize=(6,4))
        sns.boxplot(x='Diabetes_012', y=col, data=diabetes, palette='Set2')
        plt.title(f'{col} by Diabetes Status', fontsize=14)
        plt.xlabel("Diabetes Status (0=No, 1=Yes, 2=Prediabetes)")
        plt.ylabel(col)
        plt.show()
```

```python
# 7. Countplots for Categorical Variables

categorical_vars = ['Diabetes_012', 'HighBP', 'HighChol', 'Smoker', 
                    'Stroke', 'HeartDiseaseorAttack', 'Sex']

for col in categorical_vars:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=diabetes, palette='Set2')
    plt.title(f'Count of {col}', fontsize=14)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
# 8. Save Summary Stats & Correlation Matrix

numeric_summary = diabetes.describe()
numeric_summary.to_csv("eda_summary_diabetes.csv")
corr_matrix.to_csv("correlation_matrix_diabetes.csv")
print("Saved summary stats and correlation matrix as CSV files.")
```

```python
# 9. Pairplot of Selected Features

selected_features = ['BMI', 'Age', 'HighBP', 'HighChol', 'Diabetes_012']
sns.pairplot(diabetes[selected_features], hue='Diabetes_012', palette='Set2', diag_kind='kde')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```
